# 微博文本预处理程序

## 1.导入必要的包 

```python
import pandas as pd
```

## 2.构造文件路劲以及打开文件 

```python
datapath='./data/'
data = pd.read_csv(datapath+"clear_data.txt",delimiter='\t')
data.head()
```

```python
type(data)
```

## 3.将pd文件转化成列表形式，供后序程序读取和处理 

```python
prepro_txt = data["WEIBO_TXT"]
prepro_txt.head()
```

```python
#新建列表,用于承接prepro_txt中的数据
new_data = []

for i in range(len(prepro_txt)):
    new_data.append(prepro_txt[i])
```

```python
#定义一个函数，进行文本预处理
import re #导入正则表达式库
def clean_txt(new_datas):
    for i in range(len(new_datas)):#for循环对中间的每一条进行匹配和清理
        new_datas[i]=re.sub("@([\s\S]*?) ","",new_datas[i])
        new_datas[i]=re.sub("上海·上海交通大学\(闵行校区\)","",new_datas[i])
        new_datas[i]=re.sub("上海·华东师范大学\(闵行校区\)","",new_datas[i])
        new_datas[i]=re.sub("铜陵·铜陵职业技术学院","",new_datas[i])
        new_datas[i] = re.sub("\#([\S\s]*?)\#","",new_datas[i]) 
        new_datas[i] = re.sub("...展开全文","",new_datas[i])
    return new_datas
```

```python
new_datas = clean_txt(new_data)
```

```python
len(new_datas)
```

```python
for i in range(len(new_datas)):
    print(new_datas[i])
```

## 4.数据导出到txt文件中 

```python
new_clear_data = open("./data/prepro_txt.txt",'w',encoding='utf-8')
```

```python
for i in range(len(new_datas)):
    new_clear_data.write(new_datas[i]) 
```
